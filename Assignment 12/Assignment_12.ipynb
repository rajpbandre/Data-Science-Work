Use Standard Scaling (Z-score):

Preferred when data is normally distributed.

Used in algorithms that assume Gaussian distribution like logistic regression, linear regression, SVM, etc.

Min-Max Scaling:

Used when data needs to be within a specific range (0–1).

Helpful for algorithms like KNN, neural networks, or when features have different scales.

Label Encoding

Pros:

Simple and easy to implement.

Does not increase the dimensionality of the data.

Useful for ordinal variables (where order matters).

Cons:

Introduces implied ordinal relationships even when the data is categorical (e.g., color: red, blue, green → 0, 1, 2), which can mislead models.

Some algorithms may treat higher label values as having more weight or importance.


 One-Hot Encoding

Pros:

Suitable for nominal (unordered) categories.

Prevents models from assuming any ordinal relationship.

Works well with algorithms that can’t handle categorical values directly (like logistic regression or SVM).

Cons:

Increases dimensionality significantly when the number of categories is high (known as the “curse of dimensionality”).

Can lead to sparse data and increased computational cost.
